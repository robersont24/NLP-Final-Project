```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 14 14:46:23 2026

@author: taylorroberson4
"""

#Visual 1: 
    
import matplotlib.pyplot as plt
import numpy as np

# Data
challenges = [
    "Dependency Issues",
    "Inconsistent Metric Structures",
    "No Aggregation Pipeline",
    "Hardware Limits",
    "Limited Model Coverage",
    "Could Not Reproduce Fin-E5",
    "Missing Infrastructure Details",
    "No End-to-End Script"
]

severity = [4, 4, 5, 5, 4, 5, 4, 5]

# Color mapping
colors = ["#E89C5B" if s == 4 else "#E36F4E" for s in severity]

# Plot
plt.figure(figsize=(11, 6.7))
bars = plt.barh(challenges, severity, color=colors)

# Labels at end of bars
for bar, val in zip(bars, severity):
    plt.text(bar.get_width() + 0.05,
             bar.get_y() + bar.get_height() / 2,
             f"{val}",
             va='center',
             fontsize=12)

plt.title("Practical Reproducibility Challenges in FinMTEB", fontsize=18)
plt.xlabel("Severity", fontsize=14)
plt.xlim(0, 5.3)
plt.grid(axis='x', linestyle='--', alpha=0.35)

plt.tight_layout()
plt.show()

#Visual 2

# Data
tasks = [
    "Classification",
    "Clustering",
    "PairClassification",
    "Reranking",
    "Retrieval",
    "STS",
    "Summarization"
]

models = ["BGE-base", "E5-base", "MiniLM"]

scores = np.array([
    [0.641, 0.654, 0.579],
    [0.560, 0.561, 0.537],
    [0.679, 0.662, 0.583],
    [0.977, 0.976, 0.973],
    [0.619, 0.625, 0.568],
    [0.355, 0.388, 0.309],
    [0.454, 0.526, 0.364]
])

# Plot
fig, ax = plt.subplots(figsize=(10.5, 8))
im = ax.imshow(scores, aspect='auto', cmap='YlOrRd')

# Axis ticks
ax.set_xticks(np.arange(len(models)))
ax.set_xticklabels(models, fontsize=14)
ax.set_yticks(np.arange(len(tasks)))
ax.set_yticklabels(tasks, fontsize=14)

# Labels and title
ax.set_xlabel("Model", fontsize=18)
ax.set_ylabel("Task", fontsize=18)
ax.set_title("Reproduced FinMTEB Task Performance Heatmap", fontsize=22)

# Annotate cells
for i in range(scores.shape[0]):
    for j in range(scores.shape[1]):
        text_color = "white" if scores[i, j] > 0.75 else "black"
        ax.text(j, i, f"{scores[i, j]:.3f}",
                ha="center", va="center",
                color=text_color, fontsize=16, fontweight='bold')

# Colorbar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Score", fontsize=18)
cbar.ax.tick_params(labelsize=12)

plt.tight_layout()
plt.show()

#Visual 3

import pandas as pd

# Data
models = ["MiniLM", "BGE-base", "E5-base"]
means = [0.568, 0.619, 0.625]
stds  = [0.304, 0.310, 0.313]
mins  = [0.106, 0.135, 0.158]
maxs  = [0.980, 0.989, 0.987]

x = np.arange(len(models))

# Error bars based on min/max range around mean
lower_err = np.array(means) - np.array(mins)
upper_err = np.array(maxs) - np.array(means)
yerr = np.vstack([lower_err, upper_err])

fig, (ax1, ax2) = plt.subplots(
    1, 2, figsize=(18, 8),
    gridspec_kw={'width_ratios': [1.2, 1]}
)

scatter_colors = ["#8FB6D8", "#3E6FA3", "#E98A2B"]

ax1.errorbar(
    x, means, yerr=yerr,
    fmt='none', ecolor='dimgray',
    elinewidth=2.3, capsize=10
)

ax1.scatter(x, means, s=220, color=scatter_colors, zorder=3)

for i, m in enumerate(means):
    ax1.text(x[i], maxs[i] + 0.015, f"{m:.3f}",
             ha='center', va='bottom',
             fontsize=17, fontweight='bold')

ax1.set_xticks(x)
ax1.set_xticklabels(models, fontsize=18)
ax1.set_ylabel("Retrieval Score", fontsize=18)
ax1.set_title("Retrieval Consistency Across Datasets", fontsize=22)
ax1.set_ylim(0, 1.02)
ax1.grid(axis='y', linestyle='--', alpha=0.4)

ax2.axis('off')
ax2.set_title("Retrieval Consistency Summary", fontsize=22, pad=18)

summary_df = pd.DataFrame({
    "Model": models,
    "Mean": means,
    "Std": stds,
    "Min": mins,
    "Max": maxs
})

col_x = [0.05, 0.45, 0.64, 0.79, 0.93]
headers = ["Model", "Mean", "Std", "Min", "Max"]


for x_pos, header in zip(col_x, headers):
    ax2.text(x_pos, 0.88, header,
             fontsize=17, fontweight='bold',
             ha='center', va='center',
             transform=ax2.transAxes)


ax2.plot([0.05, 0.95], [0.85, 0.85], color='black', lw=1.2, transform=ax2.transAxes)


row_y = [0.72, 0.56, 0.40]
for y, (_, row) in zip(row_y, summary_df.iterrows()):
    ax2.text(col_x[0], y, row["Model"], fontsize=18, ha='left', va='center', transform=ax2.transAxes)
    ax2.text(col_x[1], y, f'{row["Mean"]:.3f}', fontsize=18, ha='center', va='center', transform=ax2.transAxes)
    ax2.text(col_x[2], y, f'{row["Std"]:.3f}', fontsize=18, ha='center', va='center', transform=ax2.transAxes)
    ax2.text(col_x[3], y, f'{row["Min"]:.3f}', fontsize=18, ha='center', va='center', transform=ax2.transAxes)
    ax2.text(col_x[4], y, f'{row["Max"]:.3f}', fontsize=18, ha='center', va='center', transform=ax2.transAxes)


for y in [0.78, 0.62, 0.46, 0.30]:
    ax2.plot([0.05, 0.95], [y, y], color='gray', lw=1, alpha=0.35, transform=ax2.transAxes)

plt.tight_layout()
plt.show()

#Visual 4

# Data
our_models = ["E5-base", "BGE-base", "MiniLM", "Fin-E5", "voyage-3-large", "bge-large-en-v1.5"]
our_ranks  = [1, 2, 3, 0, 0, 0]

paper_models = ["Fin-E5", "voyage-3-large", "bge-large-en-v1.5"]
paper_ranks  = [1, 2, 3]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14.5, 8))
fig.suptitle("Ranking Comparison: Reproduced Models vs Paper Leaders", fontsize=22, y=0.96)


colors_left = ["#ED8B2C", "#4F79A9", "#91B5CF", "#D9D9D9", "#D9D9D9", "#D9D9D9"]
bars1 = ax1.barh(our_models, our_ranks, color=colors_left)

for bar, rank in zip(bars1, our_ranks):
    if rank > 0:
        ax1.text(rank + 0.04,
                 bar.get_y() + bar.get_height()/2,
                 f"{rank}",
                 va='center', fontsize=17)

ax1.set_title("Our Reproduced Model Ranking", fontsize=20)
ax1.set_xlabel("Rank", fontsize=18)
ax1.set_xlim(0, 3.5)
ax1.grid(axis='x', linestyle='--', alpha=0.35)


colors_right = ["#ED8B2C", "#F0B471", "#4F79A9"]
bars2 = ax2.barh(paper_models, paper_ranks, color=colors_right)

for bar, rank in zip(bars2, paper_ranks):
    ax2.text(rank + 0.04,
             bar.get_y() + bar.get_height()/2,
             f"{rank}",
             va='center', fontsize=17)

ax2.set_title("Top Paper Model Ranking", fontsize=20)
ax2.set_xlabel("Rank", fontsize=18)
ax2.set_xlim(0, 3.5)
ax2.grid(axis='x', linestyle='--', alpha=0.35)

plt.tight_layout()
plt.show()
```
